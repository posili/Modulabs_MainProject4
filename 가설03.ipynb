```python
# 필요한 컬럼 선택
selected_columns = [
    'order_id',
    'customer_unique_id',
    'customer_zip_code_prefix',
    'customer_city',
    'customer_state',
    'order_item_id',
    'product_id',
    'product_category_name_english'
]

# 새로운 데이터프레임 생성
customers_orders_new= customers_orders[selected_columns]
```

지역별로 추려서 5개의 그룹으로 그룹할 예정






```python
# 주(State)별 지역 그룹화
region_mapping = {
    "North": ["AC", "AM", "AP", "PA", "RO", "RR", "TO"],
    "Northeast": ["AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"],
    "Central-West": ["DF", "GO", "MT", "MS"],
    "Southeast": ["ES", "MG", "RJ", "SP"],
    "South": ["PR", "RS", "SC"]
}

# 데이터프레임에 지역 추가
customers_orders_new["region"] = customers_orders_new["customer_state"].map(lambda x: next((k for k, v in region_mapping.items() if x in v), "Unknown"))

# 1. 지역(Region)별 인기 제품 분석
region_popular_products = customers_orders_new.groupby(['region', 'product_category_name_english']).size().reset_index(name='purchase_count')

# 각 지역에서 가장 많이 판매된 제품 카테고리 찾기
top_products_per_region = region_popular_products.sort_values(['region', 'purchase_count'], ascending=[True, False]).groupby('region').first().reset_index()

# 2. 특정 지역 고객의 구매 패턴 분석
# 고객이 구매한 제품과 지역별 인기 제품 매칭
customer_purchases = customers_orders_new.merge(top_products_per_region, on='region', suffixes=('_customer', '_region'))

# 고객이 자신의 지역에서 인기 있는 제품을 구매했는지 확인
customer_purchases['matches_region_popularity'] = customer_purchases['product_category_name_english_customer'] == customer_purchases['product_category_name_english_region']

# 지역(region)별 인기 제품을 구매한 고객 비율 계산
region_popularity_match_rate = customer_purchases.groupby('region')['matches_region_popularity'].mean().reset_index()

# 3. 지역(region)별 주문 통계량 계산
region_stats = customers_orders_new.groupby('region').agg(
    total_orders_per_region=('order_id', 'count'),  # 총 주문 개수
    unique_customers_per_region=('customer_unique_id', 'nunique'),  # 고유 고객 수
    avg_order_items_per_region=('order_item_id', 'mean'),  # 평균 주문 아이템 개수
    avg_product_category_diversity=('product_category_name_english', 'nunique')  # 제품 카테고리 다양성
).reset_index()

# 4. 기존 matches_region_popularity 데이터와 결합
region_correlation_data = region_popularity_match_rate.merge(region_stats, on='region', how='left')

# 5. 상관관계 계산
correlation_matrix = region_correlation_data.drop(columns=['region']).corr()
```

→ 인기 제품 추천 전략이 효과적인 지역을 찾는 데 유용.

2. matches_region_popularity vs unique_customers_per_region (0.419, 양의 상관관계)
→ 고객 수가 많은 지역일수록 인기 제품을 더 많이 구매하는 경향이 있음.
→ 고객이 많은 지역에서 인기 제품 추천이 효과적일 가능성 큼.

3. matches_region_popularity vs avg_order_items_per_region (-0.662, 강한 음의 상관관계)
→ 한 번에 여러 개의 제품을 구매하는 고객이 많은 지역에서는 인기 제품을 따르지 않는 경향이 강함.
→ 개별 맞춤 추천이 필요한 지역을 찾는 데 유용.

4. matches_region_popularity vs avg_product_category_diversity (-0.128, 약한 상관관계)
카테고리 다양성이 높은 지역에서는 인기 제품을 따르는 경향이 약함.
해석:
하지만 -0.128로 상관관계가 약하므로 영향이 크지 않음.
단, 카테고리 다양성이 높은 고객에게는 개별 맞춤 추천이 필요할 수 있음.

##가설검정 수행하기

가설 검정에 필요한 컬럼만 추리기

1.  order_id : 주문ID
2.   customer_unique_id : 고객 고유 ID
3. region : 고객이 거주하는 지역
3. product_category_name_english_customer : 고객이 주문한 상품 카테고리
4. product_category_name_english_region : 해당 지역의 인기 카테고리
4. matches_region_popularity : 고객이 지역 내 인기 제품을 구매했는지 여부 (True/False)



```python
test_columns=['order_id','customer_unique_id','region','product_category_name_english_customer','product_category_name_english_region','matches_region_popularity']
T_test_df=customer_purchases[test_columns]
```

```python
from scipy.stats import ttest_1samp
```

```python
# ✅ 특정 지역의 고객이 지역 내 인기 제품을 선호하는지 검정
# - matches_region_popularity 비율이 50%와 유의미하게 다른지 검증

# 1. 가설 설정
# H0: 특정 지역의 고객은 지역 내 인기 제품을 선호하지 않는다. (비율 = 50%)
# H1: 특정 지역의 고객은 지역 내 인기 제품을 선호한다. (비율 ≠ 50%)

# 2. 유의수준 설정
alpha = 0.05

# 3. 검정통계량 산출
test_mean = 0.50  # 기준 비율 (40%)
t_stat, p_value = ttest_1samp(T_test_df['matches_region_popularity'], test_mean)

# 4. 가설 기각/채택
if p_value < alpha:
    result = "귀무가설 기각 → 특정 지역의 고객은 지역 내 인기 제품을 선호한다."
else:
    result = "귀무가설 채택 → 특정 지역의 고객은 지역 내 인기 제품을 선호하지 않는다."

# 5. 결과 출력
print(f"t-통계량: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"결론: {result}")
```

- p-value = 0.0000 (< 0.05)
→ 유의수준(α = 0.05)보다 작기 때문에, 귀무가설(H₀) 기각
→ 즉, 특정 지역 고객이 해당 지역에서 인기 제품을 특별히 더 선호한다는 강력한 증거가 있음.

- t-통계량 = -424.3471
→ 매우 낮은 t-값 → 검정 기준(test_mean= 50%)보다 훨씬 낮은 값을 나타냄.
→ 실제 고객들의 인기 제품 구매율이 50%보다 훨씬 낮을 가능성이 큼.
→ 즉, 특정 지역 고객이 인기 제품을 선호하긴 하지만, 예상보다 덜 구매하고 있을 가능성이 있음.
